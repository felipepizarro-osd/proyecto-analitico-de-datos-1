```python
df = df.rename(columns={'Identificacion;Sucursal;Ciudad;Tipo de cliente;Genero;Categoria;Precio unitario;Cantidad;Impuesto 5%;Total;Fecha;Hora;Metodo de pago;cogs;Porcentaje de margen bruto;Ingresos brutos;Clasificacion': 'Header'})
df
```

```python

from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:198252021298@localhost:5432/test11')
```

```python
dfFact = pd.DataFrame(df,columns=['Identificacion',
                                   'Precio unitario',
                                  'Cantidad',
                                   'Impuesto 5%',
                                   'Total',
                                   'cogs',
                                    'Porcentaje de margen bruto',
                                    'Ingresos brutos'
                                   ])
dfFact.rename(columns={
                                    'Identificacion':'identificacion',
                                   'Precio unitario':'precio_unitario',
                                    'Cantidad':'cantidad',
                                   'Impuesto 5%':'impuesto',
                                   'Total':'total',
                                   'cogs':'cogs',
                                    'Porcentaje de margen bruto':'porcentaje_de_margen_bruto',
                                    'Ingresos brutos':'ingresos_brutos'
},inplace=True)
dfClient = pd.DataFrame(df,columns=['Identificacion',
                                   'Tipo de cliente',
                                   'Genero',
                                   'Clasificacion',
                                   ])
dfClient.rename(columns={
                                    'Identificacion':'identificacion',
                                   'Tipo de cliente':'tipo_de_cliente',
                                   'Genero':'genero',
                                   'Clasificacion':'clasificacion'
},inplace=True)
dfCiudad = pd.DataFrame(df,columns=['Identificacion',
                                   'Ciudad'])

dfCiudad.rename(columns={
                                    'Identificacion':'identificacion',
                                    'Ciudad':'nombre'
  
},inplace=True)

dfSucursal = pd.DataFrame(df,columns=['Identificacion',
                                     'Sucursal'])
dfSucursal.rename(columns={
    'Identificacion':'identificacion',
    'Sucursal':'denominacion'
},inplace=True)
dfTransaccion= pd.DataFrame(df,columns=['Identificacion',
                            'Categoria',
                            'Metodo de pago'])
dfTransaccion.rename(columns={
    'Identificacion':'identificacion',
    'Categoria':'nombre_categoria',
    'Metodo de pago':'metodo_pago'
},inplace=True)
dfTiempo = pd.DataFrame(df,columns=['Identificacion',
                        'Fecha',
                        'Hora'])
dfTiempo.rename(columns={
    'Identificacion':'identificacion',
    'Fecha':'fecha',
    'Hora':'hora'

},inplace=True)
```

```python
from sqlalchemy import text 
with engine.connect() as con:
    with open('/home/pipe/Projects/tarea_proyecto_analitico01/creation_table.sql') as file:
        query = text(file.read())
        con.execute(query)
```

```python

dfFact.to_sql('fact_table',engine,if_exists='append', index=False)
dfClient.to_sql('dimension_cliente',engine,if_exists='append', index=False)
dfCiudad.to_sql('dimension_ciudad',engine,if_exists='append', index=False)
dfSucursal.to_sql('dimension_sucursal',engine,if_exists='append', index=False)
dfTransaccion.to_sql('dimension_transaccion',engine,if_exists='append', index=False)
dfTiempo.to_sql('dimension_tiempo',engine,if_exists='append', index=False)
```
